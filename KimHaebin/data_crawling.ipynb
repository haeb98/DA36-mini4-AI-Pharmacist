```python
# Python3 샘플 코드 #
import requests
import json
import pandas as pd

url = 'http://apis.data.go.kr/1471000/DrbEasyDrugInfoService/getDrbEasyDrugList'

params = {
    'serviceKey': 'lsEb9zFwBCDA0UlPyqp7AOSWiu2gQnNKwjTvhLM/GPak4I+el0L9wrBg+npvf9jdachl66FER12MFJE+MTDs0w==',
    'pageNo': 1,  # 숫자로 유지
    'numOfRows': '100',
    'entpName': '', 
    'itemName': '', 
    'itemSeq': '', 
    'efcyQesitm': '', 
    'useMethodQesitm': '', 
    'atpnWarnQesitm': '', 
    'atpnQesitm': '', 
    'intrcQesitm': '', 
    'seQesitm': '', 
    'depositMethodQesitm': '', 
    'openDe': '', 
    'updateDe': '', 
    'type': 'json'
}

all_data = []

while True:
    response = requests.get(url, params={**params, 'pageNo': params['pageNo']})  # 현재 페이지 반영
    response = response.content.decode('utf-8')
    
    data = json.loads(response)
    if 'body' not in data or 'items' not in data['body']:  # 데이터가 없으면 중단
        break
    
    items = data['body']['items']
    if not items:  # 더 이상 데이터가 없으면 중단
        break
        
    all_data += items  # 데이터 추가
    params['pageNo'] += 1  # 페이지 번호 증가
    
df = pd.DataFrame(all_data)

# 데이터 확인
print(f"총 {len(df)}개의 데이터를 불러왔습니다.")
df.head()

# CSV 파일 저장
df.to_csv('drug_info_all_pages.csv', index=False, encoding='utf-8-sig')
print("모든 페이지 데이터를 'drug_info_all_pages.csv'로 저장했습니다.")
```
